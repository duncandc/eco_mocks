```python
from __future__ import print_function, division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Abundance Functions

## stellar mass mock

```python
# get path to mocks
import os
cwd = os.getcwd()
data_path = os.path.abspath(os.path.join(os.path.dirname(cwd), 'mocks')) + '/'
```

```python
from astropy.table import Table
mock = Table.read(data_path + 'mock_1a_00.hdf5', path='data')

Lbox = np.array([130.0,130.0,130.0])  # Vishnu box size
```

```python
from eco_mocks.galaxy_abundance_functions import Eckert_2016_phi

# stellar mass function
msample = np.logspace(7,12,1000)
phi_1 = Eckert_2016_phi(sample='RESOLVE-B SMF double')
phi_2 = Eckert_2016_phi(sample='ECO SMF double')

bins = np.arange(8,12,0.1)
bins = 10.0**bins
bin_centers = (bins[:-1]+bins[1:])/2.0

counts = np.histogram(mock['stellar_mass'], bins=bins)[0]
counts = counts/np.prod(Lbox)/0.1

fig = plt.figure()
plt.title('mock 1a')
p1, = plt.plot(bin_centers, counts, 'o')
l1, = plt.plot(msample, phi_1(msample), '-')
l2, = plt.plot(msample, phi_2(msample), '-')
plt.yscale('log')
plt.xscale('log')
plt.ylim([10**-5,0.1])
plt.xlim([10**7.5,10**12])
plt.legend([p1,l1,l2],['mock','RESOLVE-B', 'ECO'], loc='lower left', frameon=False, fontsize=10)
plt.xlabel(r'$M_{*}$')
plt.ylabel(r'$\phi(M_*)~[h^{3}{\rm Mpc}^{-3}{\rm dex}^{-1}]$')
plt.show()
fig.savefig('./figures/smf_1a.pdf', dpi=250)


# baryonic mass function
msample = np.logspace(7,12,1000)
phi_1 = Eckert_2016_phi(sample='RESOLVE-B BMF double')
phi_2 = Eckert_2016_phi(sample='ECO BMF double')

bins = np.arange(8,12,0.1)
bins = 10.0**bins
bin_centers = (bins[:-1]+bins[1:])/2.0

counts = np.histogram(mock['baryonic_mass'], bins=bins)[0]
counts = counts/np.prod(Lbox)/0.1

fig = plt.figure()
plt.title('mock 1a')
p1, = plt.plot(bin_centers, counts, 'o')
l1, = plt.plot(msample, phi_1(msample), '-')
l2, = plt.plot(msample, phi_2(msample), '-')
plt.yscale('log')
plt.xscale('log')
plt.ylim([10**-5,0.1])
plt.xlim([10**7.5,10**12])
plt.legend([p1,l1,l2],['mock','RESOLVE-B', 'ECO'], loc='lower left', frameon=False, fontsize=10)
plt.xlabel(r'$M_{\rm bary}$')
plt.ylabel(r'$\phi(M_*)~[h^{3}{\rm Mpc}^{-3}{\rm dex}^{-1}]$')
plt.show()
fig.savefig('./figures/bmf_1a.pdf', dpi=250)
```
